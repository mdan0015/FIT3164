# Transcriptomics + Proteomics + GDSC1

```python
import pandas as pd
import numpy as np
```

# Raw datas
## CCLE Transcriptomics - gene expression

```python
# Transcriptomics data
ccle_t = pd.read_csv("../Raw_files/CCLE_Transcriptomics.csv")
ccle_t = ccle_t.rename({"Unnamed: 0": "Broad_ID"}, axis = 'columns')
```

```python
ccle_t
# 1406 rows × 19222 columns
```

## CCLE Proteomics - protein levels

```python
ccle_p = pd.read_csv("../Raw_files/CCLE_Proteomics.csv")
```

```python
ccle_p = ccle_p.rename({"Unnamed: 0": "CCLE_Name"}, axis = 'columns')
ccle_p
# 899 rows × 215 columns
```

## Cell Lines Map
This dataset provides the Broad_ID (DepMap_ID), the Cancer Cell Line name, and COSMIC_ID, which is a unique identifier for tracking drug response data (to be used for joining with GSDC1 dataset later)

```python
ccle_info = pd.read_csv("../Raw_files/DepMap-2018q3-celllines.csv")

# delete unneeded columns
ccle_info = ccle_info.drop(columns=['Aliases', 'Sanger ID', 'Subtype Disease', 'Gender', 'Source'])

ccle_info
# 1673 rows × 4 columns
```

## Perform the mapping

```python
# Find matching of CCLE names between the CCLE proteonomics & mapping file

# CCLE_Name
common_values = ccle_info['CCLE_Name'].isin(ccle_p['CCLE_Name'])
match_count = common_values.sum()  # Counts True values (matches)
print(f"Number of matches: {match_count}")

# 882 / 899 rows can be matched (98%)
```

```python
# Find matching of ACH IDs between the CCLE Transcriptomics & mapping file

# Broad ID ~ ACH_ID
common_values = ccle_t['Broad_ID'].isin(ccle_info['Broad_ID'])
match_count = common_values.sum()  # Counts True values (matches)
print(f"Number of matches: {match_count}")

# 1227 / 1406 rows can be matched (87%)
```

```python
# Merge proteonomics DataFrame with the ccle info DataFrame on CCLE name column
# right merge onto ccle_info, as this data will be merged with transcriptomics data as well

ccle_p_merged = pd.merge(ccle_p, ccle_info, on=['CCLE_Name'])
ccle_multi = pd.merge(ccle_t, ccle_p_merged, on=['Broad_ID'])

#19439 columns correct amount

# Drop all rows with ANY missing values
ccle_multi = ccle_multi.dropna(axis='index')

# Final merged frame ~ multi-omics data
ccle_multi
```

```python
# Move CCLE_name, cosmic_id, primary disease to the front, after Broad_ID

# move ccle_name
col_to_move = ccle_multi['CCLE_Name']
ccle_multi.drop(labels=['CCLE_Name'], axis=1, inplace = True)
ccle_multi.insert(1, 'CCLE_Name', col_to_move)

# move cosmic_id -> join with GDSC data later
col_to_move = ccle_multi['COSMIC_ID']
ccle_multi.drop(labels=['COSMIC_ID'], axis=1, inplace = True)
ccle_multi.insert(2, 'COSMIC_ID', col_to_move)

# move primary_disease
col_to_move = ccle_multi['Primary Disease']
ccle_multi.drop(labels=['Primary Disease'], axis=1, inplace = True)
ccle_multi.insert(3, 'Primary Disease', col_to_move)
```

```python
# rename primary disease
ccle_multi = ccle_multi.rename({"Primary Disease": "cancer_type"}, axis = 'columns')

# Convert 'COSMIC_ID' column values to int64 data type
ccle_multi['COSMIC_ID'] = ccle_multi['COSMIC_ID'].apply(np.int64)
```

```python
# only retrieve breast cancer data
ccle_breast_cancer_multi = ccle_multi[ccle_multi['cancer_type']== 'Breast Cancer']
```

## Final Dataframe: Multi-omics data of 
### 1. Transcriptomics - gene expression
### 2. Proteonomics - proteomics

```python
# reset index
ccle_breast_cancer_multi.reset_index(drop=True, inplace=True)

ccle_breast_cancer_multi
```

```python
# Checking the multi-omics dataset for any repeated no. of cell lines
ccle_breast_cancer_multi['Broad_ID'].nunique()
ccle_breast_cancer_multi['Broad_ID'].count()
ccle_breast_cancer_multi['Broad_ID'].size
```

# GDSC1_IC50 Dataset  <span style="color:red">JOIN </span> 
## Drug response
The Genomics of Drug Sensitivity in Cancer (GDSC) dataset provides IC50 values for various drugs across multiple cancer cell lines. IC50 (half-maximal inhibitory concentration) is a measure of how much of a drug is needed to inhibit cell growth by 50%. This data helps researchers

### GDSC 1 and GDSC 2
GDSC 1 is the first release of the GDSC dataset, it has more number of drugs and higher number of cell lines in comparison to GDSC 2


```python
# GDSC dataset ~ drug response
gdsc1 = pd.read_csv("../Raw_files/GDSC1_IC50.csv")
# delete first column
del gdsc1['DATASET'] # GDSC1 data only
```

```python
gdsc1
# cosmic id, cell line name, drug name

# LN_IC50, AUC, RMSE, Z_SCORE
```

```python
# Trimming the dataset to only get relevant rows (Could potentially look at AUC, RMSE, Z_SCORE)

gdsc1 = gdsc1[['DRUG_ID', 'COSMIC_ID','DRUG_NAME', 'LN_IC50']]

# LN_IC50 is going to be the target feature
```

```python
gdsc1
```

### Dataset explanation
There are multiple drug responses for the same cancer cell line, i.e., a single drug may be used across multiple cell lines.

There are 402 drugs, and each drug were tested against multiple cancer cell lines. E.g., Drug_ID 1 (Erlotinib) was tested against 393 cancer cell lines

```python
# no. of drugs tested
gdsc1['DRUG_ID'].nunique()
```

```python
gdsc1['DRUG_NAME'].nunique()
```

```python
# Number of cancer cell lines the GDSC 1 data represents
gdsc1['COSMIC_ID'].nunique()
```

```python
# No. of COSMIC_ID (cancer cell line tested) for each drug
gdsc1.groupby('DRUG_ID')['COSMIC_ID'].nunique()
```

```python
# Some drug name have the same ID, these may be due to different naming conventions.
# E.g., brand name VS generic names
gdsc1.groupby('DRUG_NAME')['DRUG_ID'].nunique().loc[lambda x: x > 1]
```

# Merging multi omics breast cancer dataset with GDSC 1

```python
# Merge the gdsc data (drug response) with the multi-omics data, based on COSMIC_ID, to get a multi-omics + drug sensitivity dataset

multi_omics_gdsc1 = pd.merge(ccle_breast_cancer_multi, gdsc1, on = ['COSMIC_ID'])

# Each row represents a breast cancer cell line + its transcriptomics (gene expression), its proteomics (protein level) + the drug response (based on LN_IC50)
```

```python
# 3 columns added (drug_id, drug_name, LN_IC50)

multi_omics_gdsc1
```

```python
# There will be multiple drug responses per breast cancer cell line
multi_omics_gdsc1.groupby('Broad_ID')['DRUG_ID'].nunique()
```

```python
multi_omics_gdsc1 = multi_omics_gdsc1.dropna(axis='index')
len(multi_omics_gdsc1.index)

# No nulls in the dataset
```

```python
# move DRUG_ID to the front
col_to_move = multi_omics_gdsc1['DRUG_ID']
multi_omics_gdsc1.drop(labels=['DRUG_ID'], axis=1, inplace = True)
multi_omics_gdsc1.insert(1, 'DRUG_ID', col_to_move)


# move DRUG_NAME to the front
col_to_move = multi_omics_gdsc1['DRUG_NAME']
multi_omics_gdsc1.drop(labels=['DRUG_NAME'], axis=1, inplace = True)
multi_omics_gdsc1.insert(2, 'DRUG_NAME', col_to_move)
```

```python
# final multi omics drug response dataset
multi_omics_gdsc1

# 19,442 columns correct
```
