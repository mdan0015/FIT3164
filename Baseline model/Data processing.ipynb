## Cell Lines Map
This dataset provides the Broad_ID (DepMap_ID) and the Cancer Cell Line name

```python
ccle_info = pd.read_csv("../Dataset/DepMap-2018q3-celllines.csv")

# delete unneeded columns
ccle_info = ccle_info.drop(columns=['Aliases', 'Sanger ID', 'Subtype Disease', 'Gender', 'Source'])

ccle_info
# 1673 rows × 4 columns
```

## Perform the mapping

```python
# Find matching of CCLE names between the CCLE proteonomics & mapping file

# CCLE_Name
common_values = ccle_info['CCLE_Name'].isin(ccle_p['CCLE_Name'])
match_count = common_values.sum()  # Counts True values (matches)
print(f"Number of matches: {match_count}")

# 882 / 899 rows can be matched (98%)
```

```python
# Find matching of ACH IDs between the CCLE Transcriptomics & mapping file

# Broad ID ~ ACH_ID
common_values = ccle_t['Broad_ID'].isin(ccle_info['Broad_ID'])
match_count = common_values.sum()  # Counts True values (matches)
print(f"Number of matches: {match_count}")

# 1227 / 1406 rows can be matched (87%)
```

```python
# Merge proteonomics DataFrame with the ccle info DataFrame on CCLE name column
# right merge onto ccle_info, as this data will be merged with transcriptomics data as well

ccle_p_merged = pd.merge(ccle_p, ccle_info, on=['CCLE_Name'], how = 'right')
ccle_multi = pd.merge(ccle_t, ccle_p_merged, on=['Broad_ID'], how = 'right')

#19439 columns correct amount

# Drop all rows with ANY missing values
ccle_multi = ccle_multi.dropna(axis='index')

# Final merged frame ~ multi-omics data
ccle_multi
```

```python
# Move CCLE_name, cosmic_id, primary disease to the front, after Broad_ID

# move ccle_name
col_to_move = ccle_multi['CCLE_Name']
ccle_multi.drop(labels=['CCLE_Name'], axis=1, inplace = True)
ccle_multi.insert(1, 'CCLE_Name', col_to_move)

# move cosmic_id
col_to_move = ccle_multi['COSMIC_ID']
ccle_multi.drop(labels=['COSMIC_ID'], axis=1, inplace = True)
ccle_multi.insert(2, 'COSMIC_ID', col_to_move)

# move primary_disease
col_to_move = ccle_multi['Primary Disease']
ccle_multi.drop(labels=['Primary Disease'], axis=1, inplace = True)
ccle_multi.insert(3, 'Primary Disease', col_to_move)
```

```python
# rename primary disease
ccle_multi = ccle_multi.rename({"Primary Disease": "cancer_type"}, axis = 'columns')

# Convert 'COSMIC_ID' column values to int64 data type
ccle_multi['COSMIC_ID'] = ccle_multi['COSMIC_ID'].apply(np.int64)
```

```python
# only retrieve breast cancer data
ccle_breast_cancer_multi = ccle_multi[ccle_multi['cancer_type']== 'Breast Cancer']
```
