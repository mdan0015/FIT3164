## Cell Lines Map
This dataset provides the Broad_ID (DepMap_ID) and the Cancer Cell Line name

```python
mapping = pd.read_csv("../Dataset/DepMap-2018q3-celllines.csv")
mapping.head()
```

```python
# Find matching of CCLE names between the mapping file & the CCLE proteonomics file

common_values = mapping['CCLE_Name'].isin(ccle_p['CCLE_Name'])
match_count = common_values.sum()  # Counts True values (matches)
print(f"Number of matches: {match_count}")

# 882 / 899 rows can be matched (98%)
```

```python
# Find matching of ACH IDs between the mapping file & the CCLE Transcriptomics file

common_values = ccle_t['Broad_ID'].isin(mapping['Broad_ID'])
match_count = common_values.sum()  # Counts True values (matches)
print(f"Number of matches: {match_count}")

# 1227 / 1406 rows can be matched (87%)
```
